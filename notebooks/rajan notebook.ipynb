```python
# --- 3. Aggregate Data to Flight Level ---

# The specific key we will use for the final aggregation
flight_key = [
    'flight_number',
    'scheduled_departure_datetime_local',
    'scheduled_departure_station_code'
]

# The common key for the initial merge (using date only)
merge_key_date_only = [
    'flight_number',
    'scheduled_departure_date_local',
    'scheduled_departure_station_code'
]

# --- Corrected Passenger Aggregation ---

# Define the columns we need from the flight data: the merge key AND the specific time
columns_to_merge = [
    'flight_number',
    'scheduled_departure_date_local',
    'scheduled_departure_station_code',
    'scheduled_departure_datetime_local' # We need both date and datetime
]

# 1. MERGE FIRST to get the exact scheduled time for each passenger
pnr_data_with_datetime = pd.merge(
    pnr_flight_df,
    flight_df_cleaned[columns_to_merge], # Use the corrected list of columns
    on=merge_key_date_only,
    how='left'
)

# 2. AGGREGATE SECOND using the new combined table and the specific key
passenger_agg_df = pnr_data_with_datetime.groupby(flight_key).agg(
    total_passengers=('total_pax', 'sum'),
    child_count=('is_child_numeric', 'sum'),
    stroller_user_count=('is_stroller_user_numeric', 'sum'),
    lap_child_count=('lap_child_count', 'sum'),
    basic_economy_passengers=('basic_economy_ind', 'sum')
).reset_index()

print("‚úÖ Passenger aggregation successful!")
print(passenger_agg_df.head())


print("\nAggregating baggage data...")
# --- Corrected Baggage Aggregation ---

# 1. MERGE FIRST to get the exact scheduled time for each bag record
bag_data_with_datetime = pd.merge(
    bag_df,
    flight_df_cleaned[columns_to_merge], # Using the same columns_to_merge from before
    on=merge_key_date_only,
    how='left'
)

# 2. AGGREGATE SECOND using the new combined table and the specific key
bag_agg_df = bag_data_with_datetime.groupby(flight_key)['bag_type'].value_counts().unstack(fill_value=0).reset_index()

# Clean up column names to be more script-friendly
bag_agg_df.columns = [col.lower().replace(' ', '_').replace('(', '').replace(')', '') for col in bag_agg_df.columns]

print("‚úÖ Baggage aggregation successful!")
print(bag_agg_df.head())


# --- Corrected SSR Aggregation ---

# The specific key we will use for the final aggregation
specific_flight_key = [
    'flight_number',
    'scheduled_departure_datetime_local',
    'scheduled_departure_station_code'
]

# Create a unique list of columns needed for the link table
# This is the PNR identifier + the specific flight key
link_table_columns = [
    'record_locator',
    'pnr_creation_date',
    'flight_number',
    'scheduled_departure_datetime_local',
    'scheduled_departure_station_code'
]

# Create the link table from the enriched PNR data, ensuring unique columns
pnr_link_table = pnr_data_with_datetime[link_table_columns].drop_duplicates()

# 1. MERGE FIRST: Link SSRs to the enriched PNR data to get the specific flight key
merge_on_keys_ssr = ['record_locator', 'pnr_creation_date', 'flight_number']
remarks_with_keys = pd.merge(
    pnr_remark_df,
    pnr_link_table,
    on=merge_on_keys_ssr,
    how='left'
)

# 2. AGGREGATE SECOND: Now group by the specific flight key and count
# Drop rows where the merge couldn't find a flight key, just in case
remarks_with_keys.dropna(subset=specific_flight_key, inplace=True)
ssr_agg_df = remarks_with_keys.groupby(specific_flight_key).size().reset_index(name='ssr_count')


print("‚úÖ SSR aggregation successful!")
print(ssr_agg_df.head())

```

```python
# Start with your clean flight data as the base
final_df = flight_df_cleaned.copy()

# 1. Merge with passenger data
final_df = pd.merge(
    final_df,
    passenger_agg_df,
    on=specific_flight_key,
    how='left'
)

# 2. Merge with baggage data
final_df = pd.merge(
    final_df,
    bag_agg_df,
    on=specific_flight_key,
    how='left'
)

# 3. Merge with SSR data
final_df = pd.merge(
    final_df,
    ssr_agg_df,
    on=specific_flight_key,
    how='left'
)

# 4. Clean up missing values from the merge
# A flight with no SSRs will have NaN in 'ssr_count'. This should be 0.
# We'll apply this to all new count columns just in case.
count_columns_to_fill = [
    'total_passengers', 'child_count', 'stroller_user_count',
    'lap_child_count', 'basic_economy_passengers',
    'hot_transfer', 'origin', 'transfer',  # From bag_agg_df
    'ssr_count'
]

# Use a loop to fill NaNs with 0 for all count columns
for col in count_columns_to_fill:
    if col in final_df.columns:
        final_df[col] = final_df[col].fillna(0)


# --- Display your final, unified DataFrame ---
print("‚úÖ Merge complete! Your final analysis-ready dataset is ready.")
print("-" * 50)
print(final_df.info())
print(final_df.head())
```

```python
# a. Calculate Departure Delay in Minutes
final_df['departure_delay'] = (
    final_df['actual_departure_datetime_local'] - final_df['scheduled_departure_datetime_local']
).dt.total_seconds() / 60

# b. Calculate Passenger Load Factor
final_df['passenger_load_factor'] = final_df['total_passengers'] / final_df['total_seats']
# CORRECTED LINE: Assign the result back to the column
final_df['passenger_load_factor'] = final_df['passenger_load_factor'].fillna(0)
final_df.loc[final_df['total_seats'] == 0, 'passenger_load_factor'] = 0

# c. Calculate Ground Time Pressure
final_df['ground_time_pressure'] = final_df['scheduled_ground_time_minutes'] - final_df['minimum_turn_minutes']

# d. Calculate Transfer Bag Ratio
total_bags = final_df['origin'] + final_df['transfer'] + final_df['hot_transfer']
final_df['transfer_bag_ratio'] = final_df['transfer'] / total_bags
# CORRECTED LINE: Assign the result back to the column
final_df['transfer_bag_ratio'] = final_df['transfer_bag_ratio'].fillna(0)

print("New features engineered successfully!")
print(final_df[['departure_delay', 'passenger_load_factor', 'ground_time_pressure', 'transfer_bag_ratio']].head())
```

```python
# Interpreting Your EDA Results
# Before we build the score, let's quickly interpret what you've found:

# Significant Delays: Almost 50% of flights departing late with an average delay of ~21 minutes is a major operational challenge. This confirms the problem is real.

# Ground Time is a Key Pressure Point: Finding 621 flights scheduled with less than the minimum required ground time is a huge discovery. This is a clear, quantifiable driver of difficulty.

# Hub Complexity: An average of ~3 transfer bags for every checked bag highlights ORD's complexity as a hub. Managing transfer bags is a significant operational load.

# Passenger Load Anomaly: Your scatter plot and the correlation of -0.15 are very interesting. The plot shows some Passenger Load Factor values are greater than 1.0, which is logically impossible (you can't have more passengers than seats, excluding lap children which are handled separately).

# This "impossible" data is likely due to last-minute aircraft swaps or data entry errors. For our model, we should cap this value at 1.0. The weak negative correlation also suggests that passenger load alone isn't the primary driver of delays; other factors (like the ground time pressure you found) are likely more important.
```

```python
# Cap the passenger_load_factor at a maximum of 1.0
final_df['passenger_load_factor'] = final_df['passenger_load_factor'].clip(upper=1.0)
print("Passenger load factor cleaned.")
```

```python
from sklearn.preprocessing import MinMaxScaler

# --- Step 1: Select the features that will go into our score ---
# Based on your EDA, these are strong candidates for "difficulty"
features_for_score = [
    'ground_time_pressure',
    'passenger_load_factor',
    'transfer_bag_ratio',
    'ssr_count',
    'hot_transfer', # Number of time-sensitive transfer bags
    'child_count',
    'lap_child_count'
]

# Create a new DataFrame with just these features
score_features = final_df[features_for_score].copy()

# --- Step 2: Normalize the features ---
# We scale all features to a 0-1 range so they can be compared and combined.
# NOTE: For ground_time_pressure, lower is harder, so we'll flip it by multiplying by -1.
score_features['ground_time_pressure'] = score_features['ground_time_pressure'] * -1

scaler = MinMaxScaler()
score_features_scaled = pd.DataFrame(scaler.fit_transform(score_features), columns=score_features.columns)

# --- Step 3: Assign Weights and Calculate the Score ---
# This is where you use your insights! Weights should sum to 1.0.
# We give higher weights to factors we think are more important (like ground time).
weights = {
    'ground_time_pressure': 0.30,
    'passenger_load_factor': 0.15,
    'transfer_bag_ratio': 0.20,
    'ssr_count': 0.15,
    'hot_transfer': 0.10,
    'child_count': 0.05,
    'lap_child_count': 0.05
}

# Calculate the weighted score
final_df['difficulty_score'] = (
    score_features_scaled['ground_time_pressure'] * weights['ground_time_pressure'] +
    score_features_scaled['passenger_load_factor'] * weights['passenger_load_factor'] +
    score_features_scaled['transfer_bag_ratio'] * weights['transfer_bag_ratio'] +
    score_features_scaled['ssr_count'] * weights['ssr_count'] +
    score_features_scaled['hot_transfer'] * weights['hot_transfer'] +
    score_features_scaled['child_count'] * weights['child_count'] +
    score_features_scaled['lap_child_count'] * weights['lap_child_count']
)

# --- Step 4: Create Daily Rank and Classification ---
# This meets the deliverable requirements.

# a. Daily Rank (higher rank = more difficult)
final_df['daily_difficulty_rank'] = final_df.groupby(
    final_df['scheduled_departure_datetime_local'].dt.date
)['difficulty_score'].rank(method='dense', ascending=False)

# b. Classification into 3 Tiers
# We use quantiles (e.g., top 20%, middle 50%, bottom 30%) for classification
final_df['difficulty_class'] = pd.qcut(
    final_df['daily_difficulty_rank'],
    q=[0, 0.3, 0.8, 1.0],
    labels=['Difficult', 'Medium', 'Easy']
)


print("\n‚úÖ Flight Difficulty Score, Rank, and Class created successfully!")
# Display the final results
print(final_df[[
    'flight_number', 'scheduled_departure_datetime_local', 'difficulty_score',
    'daily_difficulty_rank', 'difficulty_class'
]].sort_values(by='difficulty_score', ascending=False).head(10))
```
